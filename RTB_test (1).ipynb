### *1. Loading the data from CSV files into the SQLITE Database*

```python
import sqlite3
import pandas as pd


# Connect to SQLite database 
conn = sqlite3.connect('rtb_database.db')
cursor = conn.cursor()

# For each CSV file read it with pandas and insert it into a new SQLite table
files_and_tables = {
    'transactions.csv': 'transactions',
    'transactions_products.csv': 'transactions_products',
    'products.csv': 'products',
    'clicks.csv': 'clicks'
}

for csv_file, table_name in files_and_tables.items():
    df = pd.read_csv(csv_file)
    df.to_sql(table_name, conn, if_exists='replace', index=False)

conn.commit()
```


### 3. Querying the Database and Exploring

###  Number of clicks on advertising banners

```python
sql_query = """ SELECT campaign_id, COUNT(*) AS number_of_clicks 
FROM clicks 
GROUP BY campaign_id;
"""

cursor.execute(sql_query)

# Fetch all rows
campaign_clicks = cursor.fetchall()

# Print each campaign's number of clicks
for campaign_id, num_clicks in campaign_clicks:
    print(f"Campaign ID {campaign_id} received {num_clicks} clicks.")
```

### Number of transactions made after our banner was clicked (so called the post click model)

```python
sql_query = """
SELECT campaign_id, COUNT(DISTINCT transaction_id) AS number_of_transactions 
FROM transactions 
GROUP BY campaign_id;
"""

cursor.execute(sql_query)

# Fetch all rows
campaign_transactions = cursor.fetchall()

# Print each campaign's number of transactions
for campaign_id, num_transactions in campaign_transactions:
    print(f"Campaign ID {campaign_id} resulted in {num_transactions} transactions.")
```

### *Advertising campaign cost which is compound of  0.01 units for a single click  and  10% out of value of each transaction*

```python
sql_query = """
SELECT 
    t.campaign_id,
    SUM(0.01 * c.number_of_clicks + 0.10 * t.transaction_value) AS total_campaign_cost
FROM 
    (SELECT campaign_id, COUNT(*) as number_of_clicks 
     FROM clicks 
     GROUP BY campaign_id) c
JOIN 
    (SELECT campaign_id, SUM(transaction_value) as transaction_value 
     FROM transactions 
     GROUP BY campaign_id) t
ON 
    c.campaign_id = t.campaign_id
GROUP BY 
    t.campaign_id;
"""


cursor.execute(sql_query)


results = cursor.fetchall()
for row in results:
    print(f"Campaign ID: {row[0]}, Total Campaign Cost: {row[1]}")

```

### *The value of all purchased products in the post-click model*

```python
sql_query = """
SELECT 
    p.brand,
    SUM(p.price) AS total_value
FROM 
    transactions t
JOIN 
    clicks c ON t.user_id = c.user_id AND t.campaign_id = c.campaign_id
JOIN 
    transactions_products tp ON t.transaction_id = tp.transaction_id
JOIN 
    products p ON tp.product_id = p.id
WHERE 
    t.date > c.date
GROUP BY 
    p.brand;
"""

cursor.execute(sql_query)

results = cursor.fetchall()
for row in results:
    print(f"Brand: {row[0]}, Total Purchase Value: {row[1]}")

# Commit the changes and close the connection
conn.commit()
conn.close()
```
